**3. Discussions:**

Color is one of the most powerful tools in data visualization, but it must be applied with purpose to enhance understanding rather than distract the audience. The effective use of color goes beyond aesthetics; it helps direct attention, establish categories, and reveal patterns that may otherwise be overlooked.

-Enhancing Clarity: Colors should be chosen to make the data more readable, not overwhelming. Overusing color can confuse the viewer, while thoughtful use highlights what is most important.

-Consistency and Professionalism: Reusing the same colors across multiple charts and dashboards helps maintain coherence, avoids misinterpretation, and conveys professionalism.

**Color and Perception**

1. Color and Perception

Human perception of color is not purely objective—it depends on context, contrast, and how our eyes and brain interpret visual signals. For example, the same shade of gray can appear lighter or darker depending on the surrounding background. This means that in visualization, color must be applied carefully, because perception can influence how data is understood.

2. Attributes of Color: Hue, Saturation, and Luminance

Hue: The type of color itself (red, blue, green, etc.).

Saturation: The intensity or purity of the color (a vivid red vs. a faded pink).

Luminance (Brightness/Value): The lightness or darkness of a color (light blue vs. dark blue).
These three attributes together determine how a color is seen and how effectively it communicates meaning in visualization.

3. Color Psychology and Color Symbolism

Colors carry psychological associations and cultural meanings:

Red → urgency, warning, or importance.

Green → growth, safety, or approval.

Blue → calm, trust, or professionalism.

Yellow → caution or attention.
While useful, designers should be cautious, since meanings can vary by culture.

4. Perceived Color

The perception of color depends on context and surroundings. For example, a mid-gray square will look lighter on a dark background and darker on a light background. This phenomenon illustrates how perceived color is not absolute, but relative.

4. Movement <br>

Movement is a strong but less commonly used preattentive attribute (especially in static charts). Our brains are highly sensitive to motion—it immediately captures attention.



*   Flicker/Animation: Blinking elements or pulsing dots draw the eye faster than color or size changes.
*   Motion Path: Direction of movement can show flow or change over time.
*   Dynamic Visualization: Used more in digital dashboards and interactive graphics rather than static charts.

**Color Schemes and Types of Data**

1. Color Schemes and Types of Data

Choosing the right color scheme depends on the type of data being visualized. The main types are:

Categorical data: Distinct groups or categories (e.g., gender, product type).

Sequential data: Ordered data where values progress from low to high (e.g., age, income, population).

Diverging data: Data with a meaningful midpoint, where deviation in both directions is important (e.g., profit vs. loss, above vs. below average).

The effectiveness of a visualization often depends on how well the color scheme matches the data type.

2. Categorical Color Schemes

Best for nominal (non-numeric) data.

Each category should have a distinct, easily distinguishable color.

Colors should not imply ordering (e.g., avoid gradients).

Example: Different regions of a country, or product categories.

3. Sequential Color Schemes

Used for ordered/numeric data that progresses from low to high.

A single hue with changes in luminance (light to dark) or saturation is often used.

Example: Population density, exam scores, rainfall.

Darker or more intense shades usually represent higher values.

4. Diverging Color Schemes

Used when data has a critical midpoint (e.g., zero, average, normal range).

Two contrasting hues show deviations in opposite directions.

Example: Positive vs. negative profit, temperature changes relative to average, political leaning.

Neutral or white/light color usually represents the midpoint.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for categorical example
categories = ["A", "B", "C", "D", "E"]
values = [10, 30, 20, 40, 25]

# Grid for sequential and diverging examples
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z_seq = np.sqrt(X**2 + Y**2)     # Sequential data (distance from origin)
Z_div = X                        # Diverging data (negative to positive)

# Create figure
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# -----------------------------
# 1. Categorical Color Scheme
# -----------------------------
axes[0].bar(categories, values, color=plt.cm.Set2.colors[:len(categories)])
axes[0].set_title("Categorical Color Scheme")

# -----------------------------
# 2. Sequential Color Scheme
# -----------------------------
seq = axes[1].imshow(Z_seq, cmap="Blues", origin="lower")
axes[1].set_title("Sequential Color Scheme")
plt.colorbar(seq, ax=axes[1], shrink=0.7)

# -----------------------------
# 3. Diverging Color Scheme
# -----------------------------
div = axes[2].imshow(Z_div, cmap="coolwarm", origin="lower")
axes[2].set_title("Diverging Color Scheme")
plt.colorbar(div, ax=axes[2], shrink=0.7)

# Show results
plt.tight_layout()
plt.show()
```

**Task 2:** Color Schemes and Types of Data

Create three different plots using sample data to demonstrate:

Categorical Scheme (different categories with distinct colors)

Sequential Scheme (values increasing in a gradient from light to dark)

Diverging Scheme (centered around a midpoint, such as negative to positive values)

Use seaborn or matplotlib colormaps (Set2, Blues, RdBu).

```python
#Code Here

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data
categories = ['Cat A', 'Cat B', 'Cat C', 'Cat D', 'Cat E']
values_cat = [12, 9, 15, 7, 11]

# Sequential data (numeric increasing values)
values_seq = np.linspace(1, 10, 5)

# Diverging data (values centered around zero)
values_div = [-5, -2, 0, 3, 6]

fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# 1. Categorical Scheme (Set2) - distinct colors for each category
palette_cat = sns.color_palette("Set2", n_colors=len(categories))
axs[0].bar(categories, values_cat, color=palette_cat)
axs[0].set_title('Categorical Scheme (Set2)')
axs[0].set_ylabel('Value')

# 2. Sequential Scheme (Blues) - gradient from light to dark
cmap_seq = plt.cm.Blues
colors_seq = cmap_seq(np.linspace(0.3, 1, len(values_seq)))  # from lighter to darker blue
axs[1].bar(range(len(values_seq)), values_seq, color=colors_seq)
axs[1].set_xticks(range(len(values_seq)))
axs[1].set_xticklabels([f'Val {i+1}' for i in range(len(values_seq))])
axs[1].set_title('Sequential Scheme (Blues)')
axs[1].set_ylabel('Value')

# 3. Diverging Scheme (RdBu) - colors diverging around zero
cmap_div = plt.cm.RdBu
# Normalize values between 0 and 1 for colormap, center zero at 0.5
norm_values = (np.array(values_div) - min(values_div)) / (max(values_div) - min(values_div))
colors_div = cmap_div(norm_values)
axs[2].bar(range(len(values_div)), values_div, color=colors_div)
axs[2].set_xticks(range(len(values_div)))
axs[2].set_xticklabels([f'Val {i+1}' for i in range(len(values_div))])
axs[2].set_title('Diverging Scheme (RdBu)')
axs[2].set_ylabel('Value')

plt.tight_layout()
plt.show()
```

**Custom Color Using the HSL Color System**

HSL Color System

The HSL model represents colors in terms of:

Hue (H): The type of color, represented as an angle on the color wheel (0–360°).

0° = Red, 120° = Green, 240° = Blue, etc.

Saturation (S): The intensity or purity of the color (0% = gray, 100% = vivid).

Luminance (L): The brightness of the color (0% = black, 50% = “normal” color, 100% = white).

Compared to RGB, the HSL model is more intuitive for humans, since we can directly control how colorful and how bright/dark a color looks.

Why use HSL in Data Visualization?

Customizability: Easily adjust brightness and vividness of the same hue.

Consistency: Keeps relationships across categories while maintaining visual balance.

Highlighting: By changing luminance or saturation, you can emphasize certain data points without changing the hue.

```python
import matplotlib.pyplot as plt
import colorsys
import numpy as np

# Function to convert HSL to RGB (Matplotlib uses RGB)
def hsl_to_rgb(h, s, l):
    return colorsys.hls_to_rgb(h, l, s)

# Create figure
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# -----------------------------
# 1. Varying Hue (different base colors)
# -----------------------------
hues = np.linspace(0, 1, 6, endpoint=False)  # 6 steps around the color wheel
colors_hue = [hsl_to_rgb(h, 1, 0.5) for h in hues]

axes[0].bar(range(len(hues)), [1]*len(hues), color=colors_hue)
axes[0].set_title("Varying Hue (0°–360°)")
axes[0].axis("off")

# -----------------------------
# 2. Varying Saturation (from grayish to vivid red)
# -----------------------------
saturations = np.linspace(0, 1, 6)
colors_sat = [hsl_to_rgb(0, s, 0.5) for s in saturations]  # Hue=0 → Red

axes[1].bar(range(len(saturations)), [1]*len(saturations), color=colors_sat)
axes[1].set_title("Varying Saturation (gray → vivid red)")
axes[1].axis("off")

# -----------------------------
# 3. Varying Luminance (dark → light green)
# -----------------------------
luminances = np.linspace(0, 1, 6)
colors_lum = [hsl_to_rgb(1/3, 1, l) for l in luminances]  # Hue=120° → Green

axes[2].bar(range(len(luminances)), [1]*len(luminances), color=colors_lum)
axes[2].set_title("Varying Luminance (dark → light green)")
axes[2].axis("off")

# -----------------------------
plt.tight_layout()
plt.show()
```

**Task 3:** Custom Color Using the HSL Color System

Write code that:

Generates 5 custom colors using the HSL system (different hues, fixed saturation/luminance).

Uses these custom colors to create a pie chart.

Prints the RGB values of each color to show the conversion.

Use colorsys or matplotlib.colors.hsv_to_rgb.

```python
#Code here

import matplotlib.pyplot as plt
import colorsys

# Parameters for HSL
saturation = 0.7
luminance = 0.5
num_colors = 5

# Generate hues evenly spaced around the color wheel
hues = [i / num_colors for i in range(num_colors)]

# Convert HSL (note: colorsys uses HLS)
def hsl_to_rgb(h, s, l):
    # colorsys uses HLS, so order is (hue, luminance, saturation)
    r, g, b = colorsys.hls_to_rgb(h, l, s)
    return (r, g, b)

colors_hsl = [hsl_to_rgb(h, saturation, luminance) for h in hues]

# Sample data for pie chart
sizes = [20, 25, 15, 30, 10]
labels = [f'Segment {i+1}' for i in range(num_colors)]

# Print RGB values (0-255 scale)
print("Custom HSL Colors (RGB values 0-255):")
for i, color in enumerate(colors_hsl):
    rgb_255 = tuple(int(c * 255) for c in color)
    print(f"{labels[i]}: {rgb_255}")

# Plot pie chart
plt.figure(figsize=(7,7))
plt.pie(sizes, labels=labels, colors=colors_hsl, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart with Custom HSL Colors')
plt.show()
```

**Common Mistakes in the Use of Color in Data
Visualization**

Common Mistakes in the Use of Color in Data Visualization

Color is powerful for highlighting, grouping, and guiding attention, but when misused, it can confuse or mislead. Below are the most frequent mistakes:

1. Unnecessary Color

Using color where it does not add value.

Example: A bar chart where every bar is a different random color, even though the variable is a single category.

Correct approach: Use a single neutral color unless highlighting is needed.

2. Excessive Color

Using too many colors at once can overwhelm the reader.

Example: Pie chart with 20+ slices, each with a unique color, making it unreadable.

Correct approach: Limit the palette to what the eye can distinguish clearly (usually 6–8 distinct hues).

3. Insufficient Contrast

Colors that are too similar in shade make it hard to differentiate categories.

Example: Two shades of light blue side by side.

Correct approach: Ensure sufficient contrast, especially for small or adjacent elements.

4. Inconsistency Across Related Charts

Using different color schemes for the same categories across multiple charts causes confusion.

Example: “Product A” is blue in one chart and red in another.

Correct approach: Keep color assignments consistent across related visuals.

5. Neglecting Colorblindness

About 8% of men and 0.5% of women have some form of color vision deficiency.

Example: Using only red and green to distinguish categories.

Correct approach: Use colorblind-friendly palettes (e.g., viridis, cividis, or Color Universal Design guidelines).

6. Not Considering the Mode of Delivery

Visualizations may look different when printed in grayscale, projected, or viewed on different devices.

Example: A subtle gradient may disappear in black-and-white print.

Correct approach: Test visuals in the medium where they will be delivered.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data
categories = ["A", "B", "C", "D"]
values = [10, 25, 17, 30]

fig, axs = plt.subplots(3, 2, figsize=(12, 10))

# 1. Unnecessary Color (Bad)
axs[0,0].bar(categories, values, color=["red", "blue", "green", "purple"])
axs[0,0].set_title("Bad: Unnecessary Colors")

# Good: One meaningful color
axs[0,1].bar(categories, values, color="steelblue")
axs[0,1].set_title("Good: Simple Color")

# 2. Excessive Color (Bad)
colors = sns.color_palette("hsv", 12)
axs[1,0].bar(range(12), np.random.randint(5,20,12), color=colors)
axs[1,0].set_title("Bad: Excessive Colors")

# Good: Limited palette
sns.barplot(x=categories, y=values, palette="Set2", ax=axs[1,1])
axs[1,1].set_title("Good: Limited Colors")

# 3. Insufficient Contrast (Bad)
axs[2,0].bar(categories, values, color="yellow")
axs[2,0].set_facecolor("white")
axs[2,0].set_title("Bad: Low Contrast")

# Good: High Contrast
axs[2,1].bar(categories, values, color="darkblue")
axs[2,1].set_facecolor("white")
axs[2,1].set_title("Good: High Contrast")

plt.tight_layout()
plt.show()
```

**Task 4:** Common Mistakes in the Use of Color in Data Visualization

Create two sets of bar charts:

1. Bad Examples

Use unnecessary colors (different random colors for each bar).

Use poor contrast (yellow on white background).

Use excessive colors (too many categories with different hues).

2. Good Examples

Use a single meaningful color.

Ensure high contrast.

Use a consistent limited palette.

Compare “bad” vs “good” side by side.

```python
#Code here

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Sample data
categories = [f'Cat {i}' for i in range(1, 8)]
values = [5, 7, 3, 6, 8, 4, 5]

fig, axs = plt.subplots(2, 3, figsize=(18, 10))

# --- BAD EXAMPLES ---

# 1. Unnecessary random colors (different random colors for each bar)
np.random.seed(42)
random_colors = np.random.rand(len(categories), 3)
axs[0,0].bar(categories, values, color=random_colors)
axs[0,0].set_title('Bad: Random Colors (Distracting)')
axs[0,0].tick_params(axis='x', rotation=45)

# 2. Poor contrast (yellow bars on white background)
axs[0,1].bar(categories, values, color='yellow')
axs[0,1].set_facecolor('white')  # white background
axs[0,1].set_title('Bad: Poor Contrast (Yellow on White)')
axs[0,1].tick_params(axis='x', rotation=45)

# 3. Excessive colors (too many hues)
palette_excessive = sns.color_palette('hls', len(categories))  # very varied hues
axs[0,2].bar(categories, values, color=palette_excessive)
axs[0,2].set_title('Bad: Excessive Colors (Too Many Hues)')
axs[0,2].tick_params(axis='x', rotation=45)

# --- GOOD EXAMPLES ---

# 1. Single meaningful color
single_color = '#1f77b4'  # a calm blue
axs[1,0].bar(categories, values, color=single_color)
axs[1,0].set_title('Good: Single Meaningful Color')
axs[1,0].tick_params(axis='x', rotation=45)

# 2. High contrast (dark bars on white background)
axs[1,1].bar(categories, values, color='navy')
axs[1,1].set_facecolor('white')
axs[1,1].set_title('Good: High Contrast (Navy on White)')
axs[1,1].tick_params(axis='x', rotation=45)

# 3. Consistent limited palette (e.g., 3 colors repeated)
limited_palette = ['#1f77b4', '#aec7e8', '#ff7f0e']  # blue and orange shades
colors_limited = [limited_palette[i % len(limited_palette)] for i in range(len(categories))]
axs[1,2].bar(categories, values, color=colors_limited)
axs[1,2].set_title('Good: Consistent Limited Palette')
axs[1,2].tick_params(axis='x', rotation=45)

# Layout adjustment and labels
for ax in axs.flat:
    ax.set_ylim(0, 10)
    ax.set_ylabel('Value')

plt.suptitle('Common Mistakes vs Good Practices in Color Use', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import colorsys
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# ==============================
# Step 1: Create dataset
# ==============================
categories = ["Category A", "Category B", "Category C", "Category D", "Category E"]
sales = [120, 80, 150, 60, 100]

# ==============================
# Step 2: Color and Perception
# ==============================
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Hue differences
colors_hue = ["red", "blue", "green", "orange", "purple"]
sns.barplot(x=categories, y=sales, hue=categories, palette=colors_hue, ax=axes[0])
if axes[0].get_legend() is not None:
    axes[0].get_legend().remove()
axes[0].set_title("Hue Differences")

# Saturation differences
colors_sat = ["#ff9999", "#ff4d4d", "#e60000", "#b30000", "#660000"]
sns.barplot(x=categories, y=sales, hue=categories, palette=colors_sat, ax=axes[1])
if axes[1].get_legend() is not None:
    axes[1].get_legend().remove()
axes[1].set_title("Saturation Differences")

# Luminance differences
colors_lum = ["#f2f2f2", "#cccccc", "#999999", "#666666", "#333333"]
sns.barplot(x=categories, y=sales, hue=categories, palette=colors_lum, ax=axes[2])
if axes[2].get_legend() is not None:
    axes[2].get_legend().remove()
axes[2].set_title("Luminance Differences")

plt.suptitle("Color and Perception", fontsize=14, y=1.05)
plt.subplots_adjust(wspace=0.4)
plt.show()

# ==============================
# Step 3: Color Schemes & Data Types
# ==============================
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# --- Categorical Scheme ---
sns.barplot(x=categories, y=sales, hue=categories, palette="Set2", ax=axes[0])
if axes[0].get_legend() is not None:
    axes[0].get_legend().remove()
axes[0].set_title("Categorical Scheme")

# --- Sequential Scheme (bar + heatmap) ---
sns.barplot(x=categories, y=sales, hue=categories, palette="Blues", ax=axes[1])
if axes[1].get_legend() is not None:
    axes[1].get_legend().remove()
axes[1].set_title("Sequential Scheme")

# Add heatmap as inset
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)

axins = inset_axes(axes[1], width="40%", height="40%", loc="upper right")
im = axins.imshow(Z, cmap="Blues", origin="lower")
plt.colorbar(im, ax=axes[1], fraction=0.046, pad=0.04)

# --- Diverging Scheme ---
avg = np.mean(sales)
colors_div = sns.diverging_palette(220, 20, as_cmap=True)
norm = plt.Normalize(min(sales), max(sales))
sm = plt.cm.ScalarMappable(cmap=colors_div, norm=norm)
colors = [sm.to_rgba(val) for val in sales]

sns.barplot(x=categories, y=sales, hue=categories, palette=colors, ax=axes[2])
if axes[2].get_legend() is not None:
    axes[2].get_legend().remove()
axes[2].set_title("Diverging Scheme")
plt.colorbar(sm, ax=axes[2], fraction=0.046, pad=0.04)

plt.suptitle("Color Schemes and Data Types", fontsize=14, y=1.05)
plt.subplots_adjust(wspace=0.3)
plt.show()

# ==============================
# Step 4: Custom Colors with HSL
# ==============================
hsl_colors = [(i/5, 0.6, 0.5) for i in range(5)]
rgb_colors = [colorsys.hls_to_rgb(h, l, s) for h, s, l in hsl_colors]
rgb_colors_255 = [(int(r*255), int(g*255), int(b*255)) for r, g, b in rgb_colors]

# Print RGB values
print("\nRGB values of chosen HSL colors:")
for cat, rgb in zip(categories, rgb_colors_255):
    print(f"{cat}: {rgb}")

# Pie chart
plt.pie(sales, labels=categories, colors=rgb_colors)
plt.title("Custom HSL Colors (Pie Chart)")
plt.show()

# ==============================
# Step 5: Common Mistakes
# ==============================
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Bad example
bad_colors = ["red", "yellow", "lime", "pink", "cyan"]
sns.barplot(x=categories, y=sales, hue=categories, palette=bad_colors, ax=axes[0])
if axes[0].get_legend() is not None:
    axes[0].get_legend().remove()
axes[0].set_title("Bad Color Use (X)")

# Good example
sns.barplot(x=categories, y=sales, hue=categories, palette="Blues", ax=axes[1])
if axes[1].get_legend() is not None:
    axes[1].get_legend().remove()
axes[1].set_title("Good Color Use (✓)")

plt.suptitle("Common Mistakes and Corrections", fontsize=14, y=1.05)
plt.subplots_adjust(wspace=0.3)
plt.show()
```

**6. Conclusion/Learnings/Analysis:**

Using color in data visualization is highly important yet not simple. Bad decisions like arbitrary colors, not enough contrast, or too many colors produce cluttered and unreadable charts. Good things like using one appropriate color, keeping good contrast strong, and limiting yourself to a small consistent collection of colors make charts readable and professional.

Color supplements what data offers in terms of narrative, but if not carefully managed, it may distract or mislead. So always choose colors deliberately by thinking about how users will see and understand the data. Then your visualizations are more effective and easier to understand.
