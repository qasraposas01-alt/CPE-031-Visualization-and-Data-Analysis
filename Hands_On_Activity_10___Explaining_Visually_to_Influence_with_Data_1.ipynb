```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [200, 250, 180, 300, 280, 350],
    "Marketing_Spend": [50, 60, 45, 70, 65, 80]
}
df = pd.DataFrame(data)

# Calculate monthly growth percentage
df["Growth_%"] = df["Sales"].pct_change() * 100

# Visualization for Investor Audience
plt.figure(figsize=(8,5))
sns.lineplot(x="Month", y="Growth_%", data=df, marker="o", color="darkgreen")

# Highlight the highest growth point
max_growth = df["Growth_%"].max()
max_month = df.loc[df["Growth_%"].idxmax(), "Month"]
plt.annotate(f"Peak Growth: {max_growth:.1f}%",
             xy=(max_month, max_growth),
             xytext=(max_month, max_growth + 5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             ha='center')

plt.title("Investor View: Monthly Sales Growth Rate", fontsize=14, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Growth (%)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Step 3. Know Your Message — Highlight What Matters

```python
# Highlight the highest sales month (preattentive color cue)
highlight = df["Sales"].idxmax()

colors = ["gray" if i != highlight else "crimson" for i in range(len(df))]
plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors)
plt.title("Highlighting the Best Month in Sales", fontsize=13)
plt.show()
```

**Task 2:** Use color or annotations to emphasize a key insight (e.g., lowest point, sudden increase).

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [200, 250, 180, 300, 280, 350],
    "Marketing_Spend": [50, 60, 45, 70, 65, 80]
}
df = pd.DataFrame(data)

# Identify the lowest sales point
min_sales = df["Sales"].min()
min_month = df.loc[df["Sales"].idxmin(), "Month"]

# Visualization emphasizing the lowest point
plt.figure(figsize=(8,5))
sns.barplot(x="Month", y="Sales", data=df, palette="Blues")

# Highlight the lowest bar in red
bars = plt.gca().patches
for bar, month in zip(bars, df["Month"]):
    if month == min_month:
        bar.set_color("crimson")  # emphasize in red

# Add annotation for the lowest point
plt.annotate(f"Lowest Sales: {min_sales}",
             xy=(df["Month"].tolist().index(min_month), min_sales),
             xytext=(df["Month"].tolist().index(min_month), min_sales + 30),
             ha='center',
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.title("Team View: Highlighting Lowest Sales Month", fontsize=14, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

Step 4. Storytelling with Charts — Beginning, Middle, End

```python
# Beginning: Context
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="skyblue")
plt.title("Beginning: Sales Started Steady", fontsize=13)
plt.show()

# Middle: Change or Challenge
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, color="orange", marker="o")
plt.title("Middle: A Drop in March Due to Low Marketing", fontsize=13)
plt.show()

# End: Resolution
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="limegreen")
plt.title("End: Recovery and Growth After March", fontsize=13)
plt.show()
```

The sales started steadily but dropped significantly in March, revealing a possible slowdown. After increasing marketing efforts in April, sales rebounded and continued rising, showing how strategic spending helped drive recovery and growth.

Step 5. Bringing It All Together — Presentation Design

```python
# Simple combined story chart
plt.figure(figsize=(8,5))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", linewidth=2, color="royalblue", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", linestyle="--", color="orange", label="Marketing Spend")

plt.fill_between(df["Month"], df["Sales"], alpha=0.1, color="royalblue")
plt.title("From Effort to Impact: Marketing Spend Drives Sales Growth", fontsize=13)
plt.legend()
plt.show()
```

```python
# DATA STORYTELLING VISUALIZATION
# Task: Visualize sales data and highlight the highest month using color + annotation

# Upload the CSV file
from google.colab import files
uploaded = files.upload()

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("sales_data_sample.csv", encoding='latin1')

# Inspect data
print("Columns available:\n", df.columns)
df.head()

# Prepare the data — group total sales per month
monthly = df.groupby('MONTH_ID')['SALES'].sum().reset_index()

# Find the month with the highest sales
max_sales = monthly['SALES'].max()
max_row = monthly.loc[monthly['SALES'].idxmax()]

# Create the bar chart
plt.figure(figsize=(8,5))
sns.barplot(x='MONTH_ID', y='SALES', data=monthly, palette='Blues')

# Highlight the highest sales month in red
bars = plt.gca().patches
for bar, month in zip(bars, monthly['MONTH_ID']):
    if month == max_row['MONTH_ID']:
        bar.set_color('crimson')

# Add annotation for the peak
plt.annotate(f"Peak Sales: {max_sales:.0f}",
             xy=(monthly['MONTH_ID'].tolist().index(max_row['MONTH_ID']), max_sales),
             xytext=(monthly['MONTH_ID'].tolist().index(max_row['MONTH_ID']), max_sales * 1.05),
             ha='center', arrowprops=dict(facecolor='black', shrink=0.05))

# Add title and labels
plt.title("Monthly Sales Overview — Highlighting the Peak Month", fontsize=14, fontweight='bold')
plt.xlabel("Month (MONTH_ID)")
plt.ylabel("Total Sales")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

# Short Data Story
print("\n📊 DATA STORY:")
print("Sales fluctuated throughout the months, reaching a clear peak in Month",
      int(max_row['MONTH_ID']), "with total sales of", f"${max_sales:,.0f}.")
print("This indicates a strong period of demand or effective marketing during that time.")
```
