**3. Discussions:**

A data dashboard is a visual interface that displays key information and performance indicators in a concise and interactive manner. Dashboards consolidate multiple data sources to provide users with insights for decision-making and performance tracking.

Principles of Effective Dashboards:

Clarity and Simplicity: Visuals should be easy to read and interpret.

Real-Time or Dynamic Data: Dashboards often pull updated data automatically.

Interactivity: Users should be able to filter, sort, or drill down into specific metrics.

Action-Oriented Insights: The dashboard should guide users toward decisions or next steps.

Taxonomies of Data Dashboards:

Data Updates â€“ Dashboards can show static (snapshot) or dynamic (real-time) data.

User Interaction â€“ Some dashboards allow filtering, searching, and selecting parameters.

Organizational Function â€“ Dashboards can serve strategic, analytical, or operational functions.

For example:

Strategic dashboards help top management monitor KPIs.

Analytical dashboards help analysts explore data trends.

Operational dashboards support real-time monitoring of systems and workflows.

```python
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
# Simulated company performance data
np.random.seed(42)
months = pd.date_range("2024-01-01", periods=12, freq="M")
regions = ["North", "South", "East", "West"]
products = ["Laptop", "Tablet", "Phone"]

data = {
    "Month": np.random.choice(months, 200),
    "Region": np.random.choice(regions, 200),
    "Product": np.random.choice(products, 200),
    "Sales": np.random.randint(5000, 20000, 200),
    "Profit": np.random.randint(500, 5000, 200)
}

df = pd.DataFrame(data)
df.head()
```

```python
# Aggregated views
sales_by_region = df.groupby("Region", as_index=False)["Sales"].sum()
sales_by_product = df.groupby("Product", as_index=False)["Sales"].sum()
sales_trend = df.groupby("Month", as_index=False)["Sales"].sum().sort_values("Month")

# Create dashboard layout
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=("Sales by Region", "Sales by Product", "Monthly Sales Trend", "Profit vs Sales"),
                    specs=[[{"type": "bar"}, {"type": "pie"}],
                           [{"colspan": 2}, None]])

# Add charts
fig.add_trace(go.Bar(x=sales_by_region["Region"], y=sales_by_region["Sales"], name="Region Sales"), row=1, col=1)
fig.add_trace(go.Pie(labels=sales_by_product["Product"], values=sales_by_product["Sales"], name="Product Sales"), row=1, col=2)
fig.add_trace(go.Scatter(x=sales_trend["Month"], y=sales_trend["Sales"], mode="lines+markers", name="Monthly Trend"), row=2, col=1)

fig.update_layout(height=700, width=950, title_text="ðŸ“Š Interactive Data Dashboard Example")
fig.show()
```

```python
# Interactive dashboard: filterable by product
fig = px.bar(df, x="Region", y="Sales", color="Product", barmode="group",
             title="Interactive Dashboard: Sales by Region and Product",
             hover_data=["Profit"])
fig.show()
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

students = ["Anna", "Ben", "Carla", "Dion", "Ella", "Finn", "Gail", "Hanz"]
subjects = ["Math", "Science", "English", "History"]
np.random.seed(42)

data = {
    "Student": np.random.choice(students, 40),
    "Subject": np.random.choice(subjects, 40),
    "Grade": np.random.randint(70, 100, 40),
    "Attendance": np.random.randint(75, 100, 40)
}

df = pd.DataFrame(data)

# --------------------------------------------
# Aggregate data
avg_grade_subject = df.groupby("Subject", as_index=False)["Grade"].mean()
avg_attendance_student = df.groupby("Student", as_index=False)["Attendance"].mean()
avg_performance = df.groupby("Student", as_index=False)["Grade"].mean()

# --------------------------------------------
# Create dashboard layout
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=("Average Grade per Subject",
                                    "Average Attendance per Student",
                                    "Overall Student Performance Trend",
                                    "Grade Distribution"),
                    specs=[[{"type": "bar"}, {"type": "bar"}],
                           [{"colspan": 2}, None]])

# Add charts
fig.add_trace(go.Bar(x=avg_grade_subject["Subject"], y=avg_grade_subject["Grade"],
                     name="Grades by Subject", marker_color='steelblue'), row=1, col=1)

fig.add_trace(go.Bar(x=avg_attendance_student["Student"], y=avg_attendance_student["Attendance"],
                     name="Attendance by Student", marker_color='darkorange'), row=1, col=2)

fig.add_trace(go.Scatter(x=avg_performance["Student"], y=avg_performance["Grade"],
                         mode='lines+markers', name="Performance Trend",
                         line=dict(color='green')), row=2, col=1)

# Update layout
fig.update_layout(height=700, width=950,
                  title_text="ðŸŽ“ Student Performance Dashboard",
                  showlegend=False)
fig.show()
```

**5. Supplementary Activity:**

Create a mini dashboard project that tracks student grades or attendance:

1. Include 3 visualizations (e.g., average grades by subject, attendance by week, and performance trend).

2. Add at least one interactive feature (dropdown or slider).

3. Explain how your dashboard helps an instructor or department make data-driven decisions.

```python
# Mini Dashboard Project: Student Grades and Attendance Tracker
# --------------------------------------------
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# --------------------------------------------
# Create manual dataset
np.random.seed(42)
students = ["Anna", "Ben", "Carla", "Dion", "Ella", "Finn", "Gail", "Hanz"]
subjects = ["Math", "Science", "English", "History"]
weeks = ["Week 1", "Week 2", "Week 3", "Week 4"]

data = {
    "Student": np.random.choice(students, 40),
    "Subject": np.random.choice(subjects, 40),
    "Week": np.random.choice(weeks, 40),
    "Grade": np.random.randint(70, 100, 40),
    "Attendance": np.random.randint(75, 100, 40)
}

df = pd.DataFrame(data)

# --------------------------------------------
# Aggregated data
avg_grade_subject = df.groupby("Subject", as_index=False)["Grade"].mean()
avg_attendance_week = df.groupby("Week", as_index=False)["Attendance"].mean()
student_performance = df.groupby("Student", as_index=False)["Grade"].mean()

# --------------------------------------------
# Create 3 visualizations using subplots
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=("Average Grades by Subject",
                                    "Average Attendance by Week",
                                    "Overall Student Performance Trend"),
                    specs=[[{"type": "bar"}, {"type": "bar"}],
                           [{"colspan": 2}, None]])

# Chart 1: Average Grades by Subject
fig.add_trace(go.Bar(x=avg_grade_subject["Subject"], y=avg_grade_subject["Grade"],
                     name="Grades by Subject", marker_color="steelblue"), row=1, col=1)

# Chart 2: Average Attendance by Week
fig.add_trace(go.Bar(x=avg_attendance_week["Week"], y=avg_attendance_week["Attendance"],
                     name="Attendance by Week", marker_color="darkorange"), row=1, col=2)

# Chart 3: Student Performance Trend
fig.add_trace(go.Scatter(x=student_performance["Student"], y=student_performance["Grade"],
                         mode="lines+markers", name="Student Trend", line=dict(color="green")),
              row=2, col=1)

# Layout settings
fig.update_layout(height=700, width=950,
                  title_text="ðŸŽ“ Mini Dashboard: Student Grades and Attendance Tracker",
                  showlegend=False)
fig.show()

# --------------------------------------------
# Interactive Feature: Dropdown to view Grades by Subject
fig2 = px.bar(df, x="Student", y="Grade", color="Subject",
              title="Interactive View: Student Grades per Subject",
              hover_data=["Attendance"])

fig2.update_layout(
    updatemenus=[
        dict(
            buttons=[
                dict(label=sub,
                     method="update",
                     args=[{"x": [df[df["Subject"] == sub]["Student"]],
                            "y": [df[df["Subject"] == sub]["Grade"]],
                            "marker.color": ["steelblue"]},
                           {"title": f"Grades for {sub}"}])
                for sub in subjects
            ],
            direction="down",
            showactive=True,
            x=0.2, y=1.15
        )
    ]
)

fig2.show()
```
