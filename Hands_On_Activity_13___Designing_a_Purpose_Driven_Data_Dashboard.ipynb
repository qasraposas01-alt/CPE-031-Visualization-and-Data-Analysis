**1. Objectives:**

**3. Discussions:**

A well-designed data dashboard is more than a collection of charts — it’s a purposeful storytelling tool that connects data to decisions.

Understanding the Purpose of the Data Dashboard:
Every dashboard must have a clear objective — whether it’s to monitor performance, analyze patterns, or inform strategic decisions. The design should align with this goal, focusing only on the most relevant data and metrics.

Considering the Needs of the Data Dashboard’s Users:
Dashboards should be designed with the user in mind. A manager may need high-level KPIs, while an analyst may need granular, filterable data. Knowing the user’s goals, skills, and workflow ensures that the dashboard supports, rather than overwhelms, their decision-making.

Data Dashboard Engineering:
Engineering a dashboard involves organizing visual elements efficiently, ensuring smooth interactivity, and maintaining data accuracy. It also includes performance considerations (like handling large datasets), responsive design, and consistent color schemes.

Step 1: Install and Import Required Libraries

```python
!pip install plotly pandas
import pandas as pd
import numpy as np
import plotly.express as px
```

Step 2: Create a Sample Dataset

We’ll simulate student performance data to design an academic performance dashboard.

```python
# Create dataset
np.random.seed(10)
students = [f"Student {i}" for i in range(1, 21)]
subjects = ["Math", "Science", "English", "Programming", "Data Analytics"]

data = {
    "Student": np.random.choice(students, 100),
    "Subject": np.random.choice(subjects, 100),
    "Score": np.random.randint(60, 100, 100),
    "Attendance (%)": np.random.randint(70, 100, 100)
}

df = pd.DataFrame(data)
df.head()
```

Step 3: Design Dashboard Components

```python
# Average Score per Subject
fig1 = px.bar(df.groupby("Subject", as_index=False)["Score"].mean(),
              x="Subject", y="Score", title="Average Score per Subject",
              color="Score", color_continuous_scale="Blues")

# Attendance vs. Score Scatter
fig2 = px.scatter(df, x="Attendance (%)", y="Score", color="Subject",
                  title="Attendance vs. Performance", hover_data=["Student"])

# Score Distribution
fig3 = px.box(df, x="Subject", y="Score", title="Score Distribution by Subject",
              color="Subject")

fig1.show()
fig2.show()
fig3.show()
```

Step 4: Combine into One Interactive Dashboard

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Aggregated data
avg_scores = df.groupby("Subject", as_index=False)["Score"].mean()
avg_attendance = df.groupby("Subject", as_index=False)["Attendance (%)"].mean()

# Dashboard layout
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=("Average Score per Subject", "Attendance vs. Score",
                                    "Score Distribution", "Average Attendance per Subject"),
                    specs=[[{"type": "bar"}, {"type": "scatter"}],
                           [{"type": "box"}, {"type": "bar"}]])

fig.add_trace(go.Bar(x=avg_scores["Subject"], y=avg_scores["Score"], name="Average Score"), row=1, col=1)
fig.add_trace(go.Scatter(x=df["Attendance (%)"], y=df["Score"], mode="markers", name="Scatter Plot"), row=1, col=2)
fig.add_trace(go.Box(x=df["Subject"], y=df["Score"], name="Score Distribution"), row=2, col=1)
fig.add_trace(go.Bar(x=avg_attendance["Subject"], y=avg_attendance["Attendance (%)"], name="Average Attendance"), row=2, col=2)

fig.update_layout(height=700, width=950, title_text="🎓 Student Performance Dashboard")
fig.show()
```

```python
# Smart Campus Energy Monitor
# --------------------------------------------
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Create sample dataset
np.random.seed(25)
areas = ["Laboratory", "Offices", "Classrooms", "Library", "Canteen"]
days = pd.date_range("2025-11-01", periods=7)

data = {
    "Area": np.random.choice(areas, 50),
    "Day": np.random.choice(days, 50),
    "Energy (kWh)": np.random.randint(200, 800, 50)
}

df_energy = pd.DataFrame(data)

# Aggregate Data
avg_energy = df_energy.groupby("Area", as_index=False)["Energy (kWh)"].mean()
daily_usage = df_energy.groupby("Day", as_index=False)["Energy (kWh)"].sum()

# Create 3 charts (bar, line, pie)
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=("Average Energy Usage by Area",
                                    "Energy Usage Over Time",
                                    "Energy Share by Area"),
                    specs=[[{"type": "bar"}, {"type": "scatter"}],
                           [{"type": "pie"}, None]])

fig.add_trace(go.Bar(x=avg_energy["Area"], y=avg_energy["Energy (kWh)"],
                     marker_color="steelblue", name="Avg Energy"), row=1, col=1)

fig.add_trace(go.Scatter(x=daily_usage["Day"], y=daily_usage["Energy (kWh)"],
                         mode="lines+markers", marker_color="darkgreen", name="Usage Trend"), row=1, col=2)

fig.add_trace(go.Pie(labels=avg_energy["Area"], values=avg_energy["Energy (kWh)"],
                     name="Energy Share"), row=2, col=1)

fig.update_layout(height=700, width=950, title_text="⚡ Smart Campus Energy Monitor Dashboard")
fig.show()
```

**Task 2:** Design for User Needs

Modify the dashboard to meet specific user needs:

If your user is a teacher, highlight underperforming students.

If your user is a student, highlight their personal progress vs. class average.

If your user is a department head, summarize subject-level averages only.

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go


np.random.seed(10)
students = [f"Student {i}" for i in range(1, 21)]
subjects = ["Math", "Science", "English", "Programming", "Data Analytics"]

data = {
    "Student": np.random.choice(students, 100),
    "Subject": np.random.choice(subjects, 100),
    "Score": np.random.randint(60, 100, 100),
    "Attendance (%)": np.random.randint(70, 100, 100)
}

df = pd.DataFrame(data)


threshold = 75
underperforming = df[df["Score"] < threshold]



# 1. Average Score per Subject
fig1 = px.bar(df.groupby("Subject", as_index=False)["Score"].mean(),
              x="Subject", y="Score", title="Average Score per Subject",
              color="Score", color_continuous_scale="Blues")

# 2. Scatter plot showing underperforming students
fig2 = px.scatter(df, x="Attendance (%)", y="Score", color="Subject",
                  hover_data=["Student"], title="Teacher View: Student Performance Overview")

# Add highlighted points for underperforming students
fig2.add_trace(go.Scatter(
    x=underperforming["Attendance (%)"], y=underperforming["Score"],
    mode="markers", name="Underperforming (<75)",
    marker=dict(size=12, color="red", symbol="x")
))

# 3. Score Distribution per Subject
fig3 = px.box(df, x="Subject", y="Score", color="Subject",
              title="Score Distribution by Subject")

# Step 5: Combine into a dashboard layout
from plotly.subplots import make_subplots

avg_scores = df.groupby("Subject", as_index=False)["Score"].mean()
avg_attendance = df.groupby("Subject", as_index=False)["Attendance (%)"].mean()

fig = make_subplots(rows=2, cols=2,
                    subplot_titles=("Average Score per Subject", "Attendance vs. Score (Highlight Underperforming)",
                                    "Score Distribution by Subject", "Average Attendance per Subject"),
                    specs=[[{"type": "bar"}, {"type": "scatter"}],
                           [{"type": "box"}, {"type": "bar"}]])

# Add charts
fig.add_trace(go.Bar(x=avg_scores["Subject"], y=avg_scores["Score"], name="Average Score"), row=1, col=1)
fig.add_trace(go.Scatter(x=df["Attendance (%)"], y=df["Score"], mode="markers", name="All Students"), row=1, col=2)
fig.add_trace(go.Scatter(x=underperforming["Attendance (%)"], y=underperforming["Score"],
                         mode="markers", name="Underperforming (<75)",
                         marker=dict(size=10, color="red", symbol="x")), row=1, col=2)
fig.add_trace(go.Box(x=df["Subject"], y=df["Score"], name="Score Distribution"), row=2, col=1)
fig.add_trace(go.Bar(x=avg_attendance["Subject"], y=avg_attendance["Attendance (%)"], name="Average Attendance"), row=2, col=2)

fig.update_layout(height=750, width=950, title_text="📊 Teacher Dashboard: Identifying Underperforming Students",
                  showlegend=True)
fig.show()

# Step 6: Dashboard Purpose Description
print("📘 Purpose of Dashboard:")
print("This teacher-focused dashboard is designed to identify underperforming students based on scores below 75.")
print("It provides visual insights into overall subject averages, score distribution, and attendance-performance relationships.")
print("By highlighting low performers in red, it allows instructors to take targeted action to improve student outcomes.")
```
